```python
data_R = data[data[60]=="R"]
data_R
```

```python
data_combined = pd.concat([data_M, data_R])
data_combined
```

```python
data_combined = data_combined.rename(columns={60: 'Category'})
```

```python
data_combined['Category'] = data_combined['Category'].astype('category')
data_combined['Category'] = data_combined['Category'].cat.codes
```

```python

data_combined['Category'] = data_combined['Category'].astype(int)
data_combined['Category'].value_counts().plot(kind='bar')
```

```python
data.groupby(60).mean()
```

```python
X = data.drop([60], axis=1)
```

```python
y = data[60]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred1=lr.predict(X_test)
```

```python
accuracy_score(y_test, y_pred1)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred2 = knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred3 = rf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred3)
```
